```python
#Code cell 1
#import pandas
import pandas as pd
data=pd.read_csv("C:/Users/esemi/Unidad 4 IT/Covid2.csv",encoding='latin1')

print(data.head())
```

**Se puede observar que en "FECHA_DEF" nos salen datos que muestran las personas que sobrevivieron al covid y los que 
si estan vivos pero son datos que no nos ayudan para la practica por lo que procedere a cambiar a valores de 
0 = muerto y 1 = vivo**

```python
data.head()
```

**Listo profe como se puede observar estas son las comlumnas que deje para realizar la practica dejando las de la actividad que son:
    "UCI", "NEUMONIA" y "Entubado"**

```python
data.info()
```

**En esta parte se procedera a transformar los datos de "FECHA_DEF" a datos numericos 0 y 1**

**Preparacion de los datos**

```python
data['Superviviente'] = data['FECHA_DEF'].apply(lambda fecha: 1 if fecha=='9999-99-99' else 0)
```

**PROFE SSALE OTRA COLUMNA QUE DICE FALLECIDO MI INTENCION ERA QUE PUSIERA VIVOS Y MUERTOS POR SEPARADO PERO NO LO LOGRE POR ESO QUEDO ESA COLUMNA**

```python
data.head()
```

**Paso 3: Aplicacion de arbol de decisiones**

```python
y_target = data["Superviviente"].values
```

```python
columns = ["INTUBADO","NEUMONIA", "UCI", "TIPO_PACIENTE","EDAD","SEXO"]
X_input = data[list(columns)].values
```

```python
from sklearn import tree
clf_train = tree.DecisionTreeClassifier(criterion="entropy", max_depth=5)
clf_train = clf_train.fit(X_input, y_target)
```

**Evaluacion del Modelo**

```python
clf_train.score(X_input,y_target)
```

**Paso 4: Visualizar el árbol con datos de si sobrevivio o no sobrevivio**

```python
clasificacion = ['Estado vivo','Estado no vivo']
```

```python
from sklearn.externals.six import StringIO
with open("covid.dot", 'w') as f:
    f = tree.export_graphviz(clf_train, out_file=f, feature_names=columns,
                          class_names=clasificacion)
```
